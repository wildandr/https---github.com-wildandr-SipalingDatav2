```python
import pandas as pd
import re

# Membaca dataset dari file CSV
df = pd.read_csv("https://raw.githubusercontent.com/wildandr/https---github.com-wildandr-SipalingDatav2/main/train.csv?token=GHSAT0AAAAAACHEG2SQHMNA5Z3C5L36DPS4ZHV4HZA")
```

```python
# Daftar kolom yang perlu diubah formatnya
kolom_konversi = ["temp", "d_point", "feels", "min_temp", "max_temp"]

# Loop melalui kolom-kolom tersebut
for kolom in kolom_konversi:
    # Membersihkan kolom dari karakter non-numerik menggunakan ekspresi reguler
    df[kolom] = df[kolom].apply(lambda x: re.sub(r'[^0-9.]', '', x))

    # Mengonversi kolom ke format yang diinginkan (2 angka desimal)
    df[kolom] = df[kolom].apply(lambda x: f'{float(x):.2f}')
```

```python
# Lakukan pengecekan jika masih ada kesalahan format pada kolom

# df = pd.read_csv("https://raw.githubusercontent.com/wildandr/https---github.com-wildandr-SipalingDatav2/main/train.csv?token=GHSAT0AAAAAACHEG2SQHMNA5Z3C5L36DPS4ZHV4HZA")

# Daftar kolom yang perlu diperiksa
kolom_periksa = ["temp", "d_point", "feels", "min_temp", "max_temp"]

# Loop melalui kolom-kolom tersebut
for kolom in kolom_periksa:
    try:
        # Mencoba mengonversi nilai-nilai dalam kolom ke tipe data float
        df[kolom] = pd.to_numeric(df[kolom], errors='raise', downcast='float')
    except ValueError as e:
        print(f"Kesalahan pada kolom '{kolom}': {e}")
```

```python
df['prssr'] = df['prssr'].apply(lambda x: re.sub(r'[^0-9]', '', str(x))[:4] if re.match(r'\d{4}', str(x)) else np.nan)
```

Mengubah Format Kolom Hum

```python
def convert_humidity(value):
    if pd.notnull(value):
        # Cek apakah nilai mengandung persen (%)
        if '%' in str(value):
            # Hapus karakter '%' dan konversi menjadi float
            return float(str(value).replace('%', '').replace(',', '.')) / 100.0
        else:
            # Konversi nilai menjadi float
            return float(str(value).replace(',', '.'))
    else:
        return value

# Terapkan fungsi konversi ke kolom "hum"
df['hum'] = df['hum'].apply(convert_humidity)
```

Membersihkan wind_spd

```python
# Membersihkan kolom 'wind_spd' dari karakter non-numerik dan mengonversi ke float
df['wind_spd'] = df['wind_spd'].str.replace(r'[^0-9.]', '', regex=True).astype(float)
```

Membersihkan wind_deg

```python
# Membersihkan kolom 'wind_deg' dari karakter non-numerik dan mengonversi ke float
df['wind_deg'] = df['wind_deg'].str.replace(r'[^0-9.]', '', regex=True).astype(float)
```

Membersihkan cloud

```python
# Membersihkan kolom 'clouds' dari karakter '%' dan mengonversi ke float
df['clouds'] = df['clouds'].str.replace('%', '', regex=False).astype(float)
```

```python
import pandas as pd
import re
import numpy as np

# Daftar kolom yang perlu diubah formatnya
kolom_konversi = ["temp", "d_point", "feels", "min_temp", "max_temp", "prssr"]

# Loop melalui kolom-kolom tersebut
for kolom in kolom_konversi:
    # Membersihkan kolom dari karakter non-numerik menggunakan ekspresi reguler
    df[kolom] = df[kolom].apply(lambda x: re.sub(r'[^0-9.]', '', str(x)))

    # Mengganti nilai non-numerik dengan NaN
    df[kolom] = df[kolom].replace('', np.nan, regex=True)

    # Mengonversi kolom ke format yang diinginkan (2 angka desimal)
    df[kolom] = df[kolom].apply(lambda x: f'{float(x):.2f}' if pd.notnull(x) else x)

# Mengubah format kolom "prssr" menjadi 4 angka paling depan saja
df['prssr'] = df['prssr'].apply(lambda x: re.sub(r'[^0-9]', '', str(x))[:4])

# Menampilkan beberapa baris pertama dari dataframe yang telah dimodifikasi
print(df.head())
```

```python
# Daftar kolom yang ingin dipindahkan ke urutan paling akhir
kolom_pindah_ke_belakang = ["sea_level", "grnd_level", "rain_3h", "snow_1h", "snow_3h", "visibility", "rain_1h"]

# Menggunakan metode reorder_columns untuk memindahkan kolom-kolom tersebut
df = df[[kolom for kolom in df.columns if kolom not in kolom_pindah_ke_belakang] + kolom_pindah_ke_belakang]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Membaca dataset dari file CSV
# df = pd.read_csv("/content/drive/MyDrive/SipalingDatav2/train.csv")

# Daftar kolom yang ingin Anda buat boxplot-nya
kolom_boxplot = ["temp", "visibility", "d_point", "feels", "min_temp", "max_temp", "prssr", "sea_level",
                 "grnd_level", "hum", "wind_spd", "wind_deg", "rain_1h", "rain_3h", "snow_1h", "snow_3h", "clouds"]

# Mengonversi kolom-kolom yang perlu diperbaiki
for kolom in kolom_boxplot:
    df[kolom] = df[kolom].apply(lambda x: re.sub(r'[^0-9.]', '', str(x)))  # Menghapus karakter non-numerik
    df[kolom] = pd.to_numeric(df[kolom], errors='coerce')  # Mengonversi ke tipe data numerik, mengabaikan kesalahan

# Mengatur jumlah kolom dan baris untuk subplot
num_cols = 3
num_rows = (len(kolom_boxplot) + num_cols - 1) // num_cols

# Membuat subplot dengan 3 kolom
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 20))

# Membuat boxplot untuk setiap kolom
for i, kolom in enumerate(kolom_boxplot):
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col]

    sns.boxplot(x=df[kolom], ax=ax)  # Membuat boxplot menggunakan Seaborn
    ax.set_title(f'Box Plot {kolom}')
    ax.set_xlabel(kolom)

# Menghapus subplot yang tidak digunakan
for i in range(len(kolom_boxplot), num_rows * num_cols):
    fig.delaxes(axes[i // num_cols, i % num_cols])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Membaca dataset dari file CSV
df = pd.read_csv("/content/drive/MyDrive/SipalingDatav2/train.csv")

# Daftar kolom yang ingin Anda periksa untuk outlier
kolom_outlier = ["temp", "visibility", "d_point", "feels", "min_temp", "max_temp", "prssr", "sea_level",
                 "grnd_level", "hum", "wind_spd", "wind_deg", "rain_1h", "rain_3h", "snow_1h", "snow_3h", "clouds"]

# Menghitung Q1 (kuartil pertama)
Q1 = df[kolom_outlier].quantile(0.25)

# Menghitung Q3 (kuartil ketiga)
Q3 = df[kolom_outlier].quantile(0.75)

# Menghitung IQR (Interquartile Range)
IQR = Q3 - Q1

# Menentukan batas atas (Upper Bound) dan batas bawah (Lower Bound)
Upper_Bound = Q3 + (1.5 * IQR)
Lower_Bound = Q1 - (1.5 * IQR)

# Mendeteksi outlier
outliers = ((df[kolom_outlier] < Lower_Bound) | (df[kolom_outlier] > Upper_Bound)).any(axis=1)

# Menampilkan outlier dalam dataset
outlier_data = df[outliers]

# Menampilkan outlier
print("Outlier Data:")
print(outlier_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca dataset dari file CSV
df = pd.read_csv("/content/drive/MyDrive/SipalingDatav2/train.csv")

# Mengatur kolom 'datetime' sebagai indeks dengan format datetime
df['datetime'] = pd.to_datetime(df['datetime'], unit='s')
df.set_index('datetime', inplace=True)

# Daftar kolom yang ingin diplot
kolom_to_plot = ["temp", "d_point", "feels", "min_temp", "max_temp"]

# Jumlah kolom per baris dalam output visualisasi
kolom_per_baris = 3

# Menghitung jumlah baris yang diperlukan
jumlah_baris = len(kolom_to_plot) // kolom_per_baris
if len(kolom_to_plot) % kolom_per_baris != 0:
    jumlah_baris += 1

# Mengatur ukuran plot
plt.figure(figsize=(15, 5 * jumlah_baris))

# Membuat visualisasi time series dengan model garis
for i, kolom in enumerate(kolom_to_plot):
    plt.subplot(jumlah_baris, kolom_per_baris, i + 1)
    plt.plot(df.index, df[kolom])
    plt.title(kolom)
    plt.xlabel('Waktu')
    plt.ylabel('Nilai')

# Menampilkan visualisasi
plt.tight_layout()
plt.show()
```
