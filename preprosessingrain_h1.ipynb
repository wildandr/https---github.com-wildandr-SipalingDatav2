```python
import pandas as pd

# Baca data dari file CSV atau sumber data lainnya
df = pd.read_csv("prep.csv")

# Konversi kolom 'datetime' ke tipe data datetime
df['datetime'] = pd.to_datetime(df['datetime'], unit='s')  # Kolom 'datetime' adalah unix timestamp

# Filter data hanya untuk bulan Desember 2010
start_date = pd.to_datetime('2010-12-01')
end_date = pd.to_datetime('2010-12-31')
december_2010_data = df[(df['datetime'] >= start_date) & (df['datetime'] <= end_date)]

# Hitung nilai rata-rata dari kolom 'rain_1h' dalam rentang Desember 2010
rain_1h_mean = december_2010_data['rain_1h'].mean()

# Mengisi nilai yang hilang dengan nilai rata-rata
december_2010_data['rain_1h'].fillna(rain_1h_mean, inplace=True)

# Plot data yang telah diisi nilai yang hilang dengan nilai rata-rata
plt.figure(figsize=(15, 5))
plt.plot(december_2010_data['datetime'], december_2010_data['rain_1h'], label='Rain_1h', color="blue")
plt.xlabel("Date")
plt.ylabel("Rain_1h")
plt.title("Rain_1h Data for December 2010 (Filled with Mean)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Baca data dari file CSV atau sumber data lainnya
df = pd.read_csv("prep.csv")

# Konversi kolom 'datetime' ke tipe data datetime
df['datetime'] = pd.to_datetime(df['datetime'], unit='s')  # Kolom 'datetime' adalah unix timestamp

# Interpolasi pada kolom 'rain_1h' untuk mengisi nilai yang hilang
df['rain_1h'].interpolate(method='linear', inplace=True)

# Simpan data yang telah diinterpolasi kembali ke "prep.csv" (menggantikan data asli)
df.to_csv("prep.csv", index=False)
```
