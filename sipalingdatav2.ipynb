```python
from autoviz import AutoViz_Class
AV = AutoViz_Class()

filename = "train.csv"
target_variable = "rain_1h"

dft = AV.AutoViz(
    filename,
    sep=",",
    depVar=target_variable,
    dfte=None,
    header=0,
    verbose=1,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
    save_plot_dir=None
)
```
